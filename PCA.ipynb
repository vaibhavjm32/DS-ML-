```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {"X" : [4, 8, 13, 7],"Y" : [11, 4, 5, 14]}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#calculate mean of x and y using numpy

mean_x = np.mean(df["X"])
mean_y = np.mean(df["Y"])
print(mean_x)
print(mean_y)
```

```python
#calaculate covariance matrix
covariance_matrix = df.cov()
print(covariance_matrix)
```

```python
type(covariance_matrix)
np_covariance = covariance_matrix.to_numpy()
```

```python
#finding eigen valeues and eigen vectors
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
l1 = eigen_values[0]
l2 = eigen_values[1]
l1v = eigen_vectors[:,0]
l2v = eigen_vectors[:,1]
print(l1)
print(l2)
print(l1v)
print(l2v)
```

```python
eigen_values
```

```python
fpc = []
for x in range(len(df["X"])):
    value = np.dot(l2v.transpose(),[[df["X"][x] - mean_x],[df["Y"][x] - mean_y]])
    fpc.append(float(value))
df["FPC"] = fpc
spc = []
for x in range(len(df["X"])):
    value = np.dot(l1v.transpose(),[[df["X"][x] - mean_x],[df["Y"][x] - mean_y]])
    spc.append(float(value))
df["SPC"] = spc
df.head()
```

```python
V = l1v
W = l2v
plt.figure(figsize=(12, 10))
plt.scatter(df["X"], df["Y"], color='black', label = '(X,Y)')
plt.xlabel("X")
plt.ylabel("Y")
plt.quiver(mean_x, mean_y, V[0], V[1], angles='xy', scale_units='xy', scale=0.1, color='black',label = 'l1v')
plt.quiver(mean_x, mean_y, -V[0], -V[1], angles='xy', scale_units='xy', scale=0.1, color='black')
plt.quiver(mean_x, mean_y, W[0], W[1], angles='xy', scale_units='xy', scale=0.1, color='b',label = 'l2v')
plt.quiver(mean_x, mean_y, -W[0], -W[1], angles='xy', scale_units='xy', scale=0.1, color='b')
plt.scatter(mean_x, mean_y, color='red', s=100, label='mean')
plt.annotate(f'Mean({mean_x},{mean_y})', (mean_x, mean_y), textcoords="offset points", xytext=(0, 16), ha='center')
for x,y in zip(df["X"], df["Y"]):
    plt.annotate(f'({x}, {y})', (x, y), textcoords="offset points", xytext=(0, 10), ha='center')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
leg = plt.legend(loc='upper right')
plt.grid()
```
